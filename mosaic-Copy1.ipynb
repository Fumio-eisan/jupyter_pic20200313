```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('012.JPG')
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
ax.imshow(img)
ax.axis("off")
```
